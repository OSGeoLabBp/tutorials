**Run RTKLIB from command prompt**

Under windows you can simply download executables from [RTKLIB demo5](https://rtklibexplorer.wordpress.com/).
In google CoLab, you need to install it which might take some time. Without going into the details, just follow the instructions below.

1.  Clone the github repository named RTKLIB demo5:

```python
!git clone https://github.com/rtklibexplorer/RTKLIB.git
```

2.  Compile and install rnx2rtkp application.

```python
%cd RTKLIB/app/consapp/rnx2rtkp/gcc
!make
!make install
```

3.  In addition to RTKLIB, a utiliy program to decompress Hatana compact RINEX files need to be downloaded, too.

```python
!wget https://terras.gsi.go.jp/ja/crx2rnx/RNXCMP_4.1.0_Linux_x86_64bit.tar.gz
```

4.  and installed, of course

```python
!tar -xvzf RNXCMP_4.1.0_Linux_x86_64bit.tar.gz
```

Then you can download RINEX observation and navigation files.

```python
!wget https://igs.bkg.bund.de/root_ftp/EUREF/obs/2022/033/BUTE00HUN_R_20220330000_01D_30S_MO.crx.gz
```

```python
!wget ftp://igs.bkg.bund.de/EUREF/BRDC/2022/033/BRDC00WRD_R_20220330000_01D_MN.rnx.gz
```

Uncompress gzip files:

```python
!gzip -d BUTE00HUN_R_20220330000_01D_30S_MO.crx.gz
!gzip -d BRDC00WRD_R_20220330000_01D_MN.rnx.gz
```

And uncompress Hatanaka compact RINEX files to RINEX files

```python
!RNXCMP_4.1.0_Linux_x86_64bit/bin/CRX2RNX BUTE00HUN_R_20220330000_01D_30S_MO.crx
```

And then you can run rnx2rtkp from command line. Output is sent to the standard output.

```python
!rnx2rtkp -p 0  BUTE00HUN_R_20220330000_01D_30S_MO.rnx BRDC00WRD_R_20220330000_01D_MN.rnx
```

You can call and run the command from a python script. There are several options,, the first one would be to apply subprocess module. A very simple example:

```python
import subprocess
subprocess.call(["rnx2rtkp", "-p", "0", "-o", "BUTE00HUN_R_20220330000_01D_30S_MO.pos", "BUTE00HUN_R_20220330000_01D_30S_MO.rnx", "BRDC00WRD_R_20220330000_01D_MN.rnx"])
```
